- we will find a score which is the weighted rating formula as a matric for sorting done by the recomendation system
- WR=(v/(v+m) * R) + (m/(m+v) * C)

```python
m= RS_df["vote_count"].quantile(0.30)
m= RS_df["vote_count"].quantile(0.30)
m
```

```python
C=RS_df["vote_average"].mean()
```

```python
C
```

- we are considering movies with time duration in the range of 45-300 minutes
- vote count should be>= to m to be in list

```python
RS_df1=RS_df[(RS_df["runtime"]>=45) & (RS_df["runtime"]<=300)]

RS_df1=RS_df1[RS_df1["vote_count"]>=m]
```

```python
RS_df1["original_language"].unique()
```

- To sort malayalam movies only on the basis of score

```python
RS_df2=RS_df1[RS_df1["original_language"]=="ml"]
```

```python
RS_df2
```

```python
#fn to calculate the imdb WR for each movie
def weigh_rating(x,m=m,C=C):
    v=x["vote_count"]
    R=x["vote_average"]
    #compute WR
    return(((v/v+m)*R) + ((m/m+v)*C))
```

```python
RS_df2["score"]= RS_df2.apply(weigh_rating,axis=1)
```

```python
RS_df2
```

- sorting

```python
RS_df2.sort_values(by="score", ascending=False, inplace=True)
```

```python
RS_df2[["title","vote_average","original_language","runtime","score"]]
```
